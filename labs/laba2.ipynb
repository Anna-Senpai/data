```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel("D:\\papaleftme\\rollingsales_bronx.xls", skiprows=4)
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=0,inplace=True)
data
```

```python
sb.set(rc={"figure.figsize": (7,9)})
sb.countplot(data=data,
             y="NEIGHBORHOOD",
             order= data["NEIGHBORHOOD"].value_counts().index).set(title="sales on district")
```

```python
sb.set(rc={"figure.figsize": (7,9)})
sb.countplot(data=data,
             y="BUILDING CLASS CATEGORY",
             order= data["BUILDING CLASS CATEGORY"].value_counts().index).set()
```

```python
order = data.groupby(["NEIGHBORHOOD"])["RESIDENTIAL UNITS"].mean().sort_values().index
sb.barplot(y='NEIGHBORHOOD',
            x='RESIDENTIAL UNITS',
            order=order,
            data=data,
            errorbar=None)
```

```python
order = data.groupby(["NEIGHBORHOOD"])["COMMERCIAL UNITS"].mean().sort_values().index
sb.barplot(y='NEIGHBORHOOD',
            x='COMMERCIAL UNITS',
            order=order,
            data=data,
            errorbar=None)
```

```python
formated_data = data[(data['TOTAL UNITS'] > 0) & (data['LAND SQUARE FEET'] > 0)]
formated_data
```

```python
data2 = formated_data[["SALE\nPRICE", "YEAR BUILT"]][formated_data["SALE\nPRICE"] > 1000000]
data2.groupby("YEAR BUILT").sum().plot(
    y="SALE\nPRICE")
```

```python
formated_data.groupby([formated_data["SALE DATE"].dt.year,
                       formated_data["SALE DATE"].dt.month])[["SALE\nPRICE"]].sum().plot(kind="bar")
```

```python
formated_data = formated_data.dropna(axis='columns')
formated_data
```

```python
formated_data["price per ft."] = formated_data["SALE\nPRICE"]/formated_data['LAND SQUARE FEET']
formated_data
```

```python
formated_data.groupby([formated_data["SALE DATE"].dt.year,
                       formated_data["SALE DATE"].dt.month])[["price per ft."]].mean().plot(kind="line")
```

```python
formated_data.groupby([formated_data["SALE DATE"].dt.year,
                       formated_data["SALE DATE"].dt.month])[["SALE\nPRICE"]].mean().plot(kind="bar")
```

```python
formated_data[formated_data["SALE\nPRICE"] > 500000].groupby([formated_data["SALE DATE"].dt.year,
                                                              formated_data["SALE DATE"].dt.month])[["SALE\nPRICE"]].count().plot(kind="bar")
```
